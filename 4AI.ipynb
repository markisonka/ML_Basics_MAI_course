Функция Himmelblau

```python
def Himmelblau(x, y): # моя функция
  return (((x**2+y-11)**2) + (((x+y**2-7)**2)))
```

```python
X=np.linspace(-6,6)
Y=np.linspace(-6,6)

x,y=np.meshgrid(X,Y)
F=Himmelblau(x,y)

fig =plt.figure(figsize=(9,9))
ax=plt.axes(projection='3d')
ax.contour3D(x,y,F,450)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('F')
ax.set_title('Himmelblau Function')
ax.view_init(50,50)

#plt.contour(x,y,F,15)
plt.show()
```

```python
def objective(trial):
    x = trial.suggest_float("x", -10, 10) #можно оптимизировать категориальные параметры
    y = trial.suggest_float("x", -10, 10)
    return (((x**2+y-11)**2) + (((x+y**2-7)**2)))
```

```python
study = optuna.create_study() #оптимизатор
study.optimize(objective, n_trials = 100)
```

```python
best_params = study.best_params
found_x = best_params["x"]
print("Found x: {}, Himmelblau : {}".format(found_x, (found_x - 2) ** 2))
```

Функция Бута





```python

def booth(x, y):
    return (x+2*y -7)*(x+2*y -7) + (2*x +y -5)*(2*x +y -5)
```

```python
X=np.linspace(-6,6)
Y=np.linspace(-6,6)

x,y=np.meshgrid(X,Y)
F=booth(x,y)

fig =plt.figure(figsize=(9,9))
ax=plt.axes(projection='3d')
ax.contour3D(x,y,F,450)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('F')
ax.set_title("Booth's Function")
ax.view_init(50,50)

#plt.contour(x,y,F,15)
plt.show()
```

```python
def obj(trial): # функц бута
    x = trial.suggest_float("x", -10, 10) #можно оптимизировать категориальные параметры
    y = trial.suggest_float("x", -10, 10)
    return (x+2*y -7)*(x+2*y -7) + (2*x +y -5)*(2*x +y -5)
```

```python
study = optuna.create_study() #оптимизатор
study.optimize(obj, n_trials = 100)
```

```python
best_params = study.best_params
found_x = best_params["x"]
print("Found x: {}, Booths : {}".format(found_x, (found_x - 2) ** 2))
```
